# **PROJECT ON NLP**

These are the given steps, which are followed for the execution of code and get the desired output.

## 1. Import the required libraries
With the help of the predefined libraries i.e. $gensim.summarization$, we will import $keywords$ which can be used for direct implementation of certain NLP basic steps.

```
from gensim.summarization import keywords       
import pandas as pd
```

## 2. Read the given dataset
With the help of **pandas** which we imported earlier, we will import the given excel dataset i.e. **NLP Dataset**, with the help of its $read\_excel()$ function. Also display the dataset, for assurance.

```
nlp = pd.read_excel("NLP Dataset.xls")
display(nlp)
```

## 3. Remove the NA or NaN values
The removal of NA and NaN values is a major prospect, as our applied algorithm also considers the NA values and thus, our desired output will vary.


*   So, to drop the NA values from our dataset (especially, the column containing data), we will use $dropna()$ function.
*   $subset$ attribute is used to perform the function on the specified column (i.e. Functional/Technical).
*   $inplace=True$ attribute is used to return none, after removing NA values.
*   $.head(50)$ is used to display the top 50 entries of the resulting dataset.








```
nlp.dropna(subset=['Functional/Technical'], inplace=True)
nlp.head(50)
```

## 4. Convertion of given column of dataframe of list
We need to convert the column of dataset to a list for further evaluations. This can be done by, $.tolist()$.

```
nlp_list = nlp['Functional/Technical'].tolist()
```

## 5. Keywords extraction from the given list
Now, we will perform the keyword extraction, through an already specified python library (i.e. keywords), which is used to display the keywords in a given text. 

The $keywords$ attribute includes:

*   Stopwords Removal
*   Tokenization
*   Stemming
*   and keyword extraction with the help of Tf-idf





```
nlp_out = []      #creates an empty list
for sen in nlp_list:    #selects a sentence from the list
  sen = str(sen)      #converts the sentence to string value
  nlp_out.append(keywords(sen, words=2))      #finds the appropriate keywords with no. of keywords as specified in 'words' argument, and appends to the empty list.
```

## 6. Appending the keywords list to the main dataframe
To get the desired output at our main dataframe, we will append our output list with the main dataframe (i.e. NLP Dataset).

```
nlp['Output'] = nlp_out     #the output column will be named as 'Output'
```

## 7. Displaying dataframe with desired output
And finally, we will display the dataset.

```
display(nlp)
```

```
nlp = nlp[['Functional/Technical', 'Output']]
new = nlp["Output"].str.split("\n", n = 1, expand = True)
nlp["First Keyword"]= new[0]
nlp["Second Keyword"]= new[1]
nlp.drop(columns =["Output"], inplace = True)
nlp.head(50)
```

```
nlp.to_excel("NLP output.xls", index=False)
```
